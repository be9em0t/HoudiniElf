```python
import ipywidgets as widgets
from IPython.display import display

preview_toggle = widgets.Checkbox(
    value=True,
    description="Enable preview"
)

run_button = widgets.Button(
    description="Run processing"
)

output = widgets.Output()

def on_run_clicked(b):
    with output:
        output.clear_output()
        print("Preview:", preview_toggle.value)
        print("Bad rows:", 12)
        print("Spaces fixed:", 34)

run_button.on_click(on_run_clicked)

display(preview_toggle, run_button, output)
```

```python
stats = widgets.HTML()

def update_stats(bad_rows, fixed_spaces):
    stats.value = f"""
    <b>Bad rows:</b> {bad_rows}<br>
    <b>Spaces fixed:</b> {fixed_spaces}
    """

update_stats(12, 34)
display(stats)
```

```python
# File dialog.
# Native, depending on OS
import platform
import subprocess

def pick_file():
    system = platform.system()

    if system == "Darwin":  # macOS
        script = '''
        set theFile to choose file
        POSIX path of theFile
        '''
        result = subprocess.run(
            ["osascript", "-e", script],
            capture_output=True,
            text=True
        )
        return result.stdout.strip()

    elif system == "Windows":
        ps_script = r'''
        Add-Type -AssemblyName System.Windows.Forms
        $ofd = New-Object System.Windows.Forms.OpenFileDialog
        $ofd.Filter = "All files (*.*)|*.*"
        if ($ofd.ShowDialog() -eq "OK") { Write-Output $ofd.FileName }
        '''
        result = subprocess.run(
            ["powershell", "-NoProfile", "-Command", ps_script],
            capture_output=True,
            text=True
        )
        return result.stdout.strip()

    else:
        raise NotImplementedError("No native file dialog for this OS")

# Usage:
file_path = pick_file()
file_path
```

```python
# Processing options UI: checkboxes + custom prompt
import ipywidgets as widgets
from IPython.display import display

chk_preview = widgets.Checkbox(value=True, description="Enable preview")
chk_fix_nbsp = widgets.Checkbox(value=True, description="Fix NBSPs")
chk_transliterate = widgets.Checkbox(value=False, description="Transliterate")
chk_normalize = widgets.Checkbox(value=True, description="Normalize")

custom_prompt = widgets.Textarea(
    value="",
    placeholder="Custom instructions for the AI agent (optional)",
    description="Custom prompt:",
    layout=widgets.Layout(width='100%'),
    rows=3
)

apply_options = widgets.Button(description="Apply options")
options_out = widgets.Output()

def on_apply_clicked(b):
    with options_out:
        options_out.clear_output()
        print("Applied options:")
        print("  Preview:", chk_preview.value)
        print("  Fix NBSPs:", chk_fix_nbsp.value)
        print("  Transliterate:", chk_transliterate.value)
        print("  Normalize:", chk_normalize.value)
        summary = custom_prompt.value if len(custom_prompt.value) < 200 else custom_prompt.value[:200] + "..."
        print("  Custom prompt (preview):", repr(summary))

apply_options.on_click(on_apply_clicked)

options_widget = widgets.VBox([
    widgets.Label("Processing options"),
    widgets.HBox([chk_preview, chk_fix_nbsp, chk_transliterate, chk_normalize]),
    custom_prompt,
    apply_options,
    options_out
])

display(options_widget)
```
